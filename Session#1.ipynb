## <font color= 'blus'>CPU, GPU & others

> Central Process Unit, consisting of billions of transistors.
>
> Executes commands, program instructions.
>
> Different kinds of processing units in a computer :
>> **CPUs** are for general purpose use. It is the computer's processor.
>>
>> **GPUs** have a specialized use: graphics rendering. It is the processor that equips the graphics cards.
>>
>> **APU**: Accelerated Process Unit these are CPUs that also integrate a GPU component.
>>
>> **SoC** : System On Chip: equips smartphones.
>>
> **GPU computing**: used in conjunction with a CPU in ML applications, this architecture speeds up calculations.
>
> Such computing takes advantage of the massively parallel architecture of the GPU to process DL pbs, the latter lending themselves well to parallelization.
>
> A GPU allows massive data processing, which a CPU does not. GPU bandwidth is much more important than a CPU.

### <font color= 'blus'>Creating virtual environment

>The `Python` language is distributed in versions. 
>
> One of the reasons is that the Python packages evolve and their anarchic installation would lead to the decoherence of the language leading to runtime errors of applications developed in Python.
>
> Anaconda allows you to create virtual environments to develop applications compatible with a given `Python` version.
>>	- `conda create -n dsaib-3.9 python=3.9`
>
> Once created virtual environemnt may be activated : 
>>	- `conda activate dsaib-3.9`
>
> A prompt looking like `(dsaib-3.9)` appears on command line
>
> From now, all install packages will be compliant with this Python version, without the requirement to specify the package version at installation stage.
>
> To deactivate virtual environment :
>>	- `conda deactivate`
